# Exploratory Data Analysis - BigMart Sales Prediction

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('train_v9rqX0R.csv')

# Shape and first look
print(train.shape)
train.head()
```

```python
# Missing values
train.isnull().sum()
```

```python
# Target distribution
sns.histplot(train['Item_Outlet_Sales'], kde=True)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(train.corr(), annot=True, cmap='coolwarm')
plt.show()
```
